```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# plotting on heat map
plt.figure(figsize=(25, 25))
sns.heatmap(df.isnull())
```

```python
"""
since in some columns data is null just delete them
we are checking how much percentage of data is missing in a column
In a column divide no. of null values by no. of rows 
"""
(df.isnull().sum() / df.shape[0]) * 100
```

```python
#  we are dropping columns in which missing data exceeds 17 %
null_var = (df.isnull().sum() / df.shape[0]) * 100
drop_columns = null_var[null_var > 17].keys()
drop_columns
```

```python
df2_drop_clm = df.drop(columns=drop_columns)
```

```python
df2_drop_clm.shape
```

```python
# plotting on a heat map
sns.heatmap(df2_drop_clm.isnull())
```

```python
# deleting rows which are containing null values
df3_drop_rows = df2_drop_clm.dropna()
```

```python
df3_drop_rows.shape
```

```python
# plotting on a heat map
sns.heatmap(df3_drop_rows.isnull())
```

```python
"""
From heatmap we can derive an conclusion that there are no null values in our data set
"""
df3_drop_rows.isnull().sum().sum()
```

```python
# selecting the columns which are numerical
df3_drop_rows.select_dtypes(include=['int64', 'float64']).columns
```

```python
# distplot of original data frame
sns.distplot(df['MSSubClass'])
```

```python
# distplot of finally obtained data frame
sns.distplot(df3_drop_rows['MSSubClass'])
```

```python
# overlapping both the distplot
sns.distplot(df['MSSubClass'])
sns.distplot(df3_drop_rows['MSSubClass'])
```

This implies whatever deletion we have done will not disturb the accuracy of ML algorithm since both are almost overlapping

Now this is only for one numerical column, but we want to see the plot for other numerical column

```python
# all numeric columns
num_var = ['MSSubClass', 'LotArea', 'OverallQual', 'OverallCond',
       'YearBuilt', 'YearRemodAdd', 'MasVnrArea', 'BsmtFinSF1', 'BsmtFinSF2',
       'BsmtUnfSF', 'TotalBsmtSF', '1stFlrSF', '2ndFlrSF', 'LowQualFinSF',
       'GrLivArea', 'BsmtFullBath', 'BsmtHalfBath', 'FullBath', 'HalfBath',
       'BedroomAbvGr', 'KitchenAbvGr', 'TotRmsAbvGrd', 'Fireplaces',
       'GarageYrBlt', 'GarageCars', 'GarageArea', 'WoodDeckSF', 'OpenPorchSF',
       'EnclosedPorch', '3SsnPorch', 'ScreenPorch', 'PoolArea', 'MiscVal',
       'MoSold', 'YrSold', 'SalePrice']
plt.figure(figsize=(25,25))
"""
we have 36 features in total
in subplot we have 9 x 4
enumerate function gives index and value from the list
"""
for i, var in enumerate(num_var) :
    plt.subplot(9, 4, i + 1)
    sns.distplot(df[var])
    sns.distplot(df3_drop_rows[var])
```

we can see from the distribution that final data and original data are mostly overlapping with each other. We can conclude that final data is accurate

# Checking distribution of categorical data

```python
# selecting the columns which are categorical
df3_drop_rows.select_dtypes(include=['object']).columns
```

```python
# frequency of all categories in 'MSZoning' column in original dataset
df['MSZoning'].value_counts()
```

```python
# calculating percentage of all categories in 'MSZoning' column in original data set
(df['MSZoning'].value_counts() / df.shape[0]) * 100
```

```python
# calculating percentage of all categories in 'MSZoning' column in refined data set
(df3_drop_rows['MSZoning'].value_counts() / df3_drop_rows.shape[0]) * 100
```

```python
pd.concat([df['MSZoning'].value_counts() / df.shape[0] * 100,
          df3_drop_rows['MSZoning'].value_counts() / df3_drop_rows.shape[0] * 100],
          axis=1,
          keys=["MSZoning_original", "MSZoning_refined"] )
```

```python
# seeing the same for other categorical values
def categorical_variable_distribution(var) :
    return pd.concat([df[var].value_counts() / df.shape[0] * 100,
                      df3_drop_rows[var].value_counts() / df3_drop_rows.shape[0] * 100],
                      axis=1,
                      keys=[var + "_original", var + "_refined"])
```

```python
categorical_variable_distribution('MSZoning')
```

```python
# all categorical variables
categorical_variable = ['MSZoning', 'Street', 'LotShape', 'LandContour', 'Utilities',
       'LotConfig', 'LandSlope', 'Neighborhood', 'Condition1', 'Condition2',
       'BldgType', 'HouseStyle', 'RoofStyle', 'RoofMatl', 'Exterior1st',
       'Exterior2nd', 'ExterQual', 'ExterCond', 'Foundation', 'BsmtQual',
       'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'Heating',
       'HeatingQC', 'CentralAir', 'Electrical', 'KitchenQual', 'Functional',
       'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond', 'PavedDrive',
       'SaleType', 'SaleCondition']
for i in categorical_variable :
    print(categorical_variable_distribution(i))
```
