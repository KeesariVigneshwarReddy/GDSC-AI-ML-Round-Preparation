```python
"""
since in some columns data is null just delete them
we are checking how much percentage of data is missing in a column
In a column divide no. of null values by no. of rows 
"""
(df.isnull().sum() / df.shape[0]) * 100
```

```python
#  we are dropping columns in which missing data exceeds 20 %
null_var = (df.isnull().sum() / df.shape[0]) * 100
drop_columns = null_var[null_var > 20].keys()
print(drop_columns)
df2_drop_clm = df.drop(columns=drop_columns)
df2_drop_clm.shape
```

```python
# numerical columns
df3_num = df2_drop_clm.select_dtypes(include=['int64', 'float64'])
df3_num.head()
```

```python
# checking the presence of missing values in heatmap
# plotting on a heat map
sns.heatmap(df3_num.isnull())
```

```python
# we are gonna extract the rows which contain null values
df3_num[df3_num.isnull().any(axis = 1)]
```

```python
df3_num.isnull().sum()
```

```python
# we have a list in which columns consisting of null values are present
missing_num_var = [i for i in df3_num.columns if df3_num[i].isnull().sum() > 0]
missing_num_var
```

```python
"""
we are going to fill missing values with either mean or median.
To choose between the two, we should take a look at distribution
"""
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(missing_num_var):
    plt.subplot(2,2, i + 1)
    sns.distplot(df3_num[var])
```

# Fill with mean value

```python
df4_num_mean = df3_num.fillna(df3_num.mean())
df4_num_mean.isnull().sum().sum()
```

```python
# overlapping of distplot before and after
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(missing_num_var):
    plt.subplot(2,2, i + 1)
    sns.distplot(df3_num[var])
    sns.distplot(df4_num_mean[var])
    plt.legend()
```

# Fill with median value

```python
df5_num_median = df3_num.fillna(df3_num.median())
# overlapping of distplot before and after
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(missing_num_var):
    plt.subplot(2,2, i + 1)
    sns.distplot(df3_num[var], hist = False)
    sns.distplot(df4_num_mean[var], hist = False)
    sns.distplot(df5_num_median[var], hist = False)
    plt.legend()
```

We will get greater difference in mean median if there are many outliers

```python
# for outliers
for i, var in enumerate(missing_num_var):
    plt.figure(figsize=(10,10))
    plt.subplot(3, 1, 1)
    sns.boxplot(df[var])
    plt.subplot(3, 1, 2)
    sns.boxplot(df4_num_mean[var])
    plt.subplot(3, 1, 3)
    sns.boxplot(df5_num_median[var])
```

There are two horizontal lines minimum and maximum. The other dots are outliers.
These outliers impact accuracy of ML algorithm.

```python
df_concat = pd.concat([df3_num[missing_num_var], df4_num_mean[missing_num_var], df5_num_median[missing_num_var]], axis = 1)
```
