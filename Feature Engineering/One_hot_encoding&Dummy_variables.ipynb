# Dummy Variables and one hot encoding

```python
# importing libraries
import pandas as pd
```

```python
# loading data set
tips_df = pd.read_csv("tips.csv")
tips_df
```

```python
# creating dummy data frame from tips_df
dummy_df = pd.get_dummies(tips_df)
dummy_df
"""
Categorical variables are selected from original data frame. New variables 
are created on the basis of number of classes in that specific variable.

For example, sex variable consists of Female and Male
             In dummy data frame, 2 columns are added
"""
```

```python
"""
we do not need an unnecessary column 
reduce no. of columns to k - 1
k = no. of classes in column
"""
pd.get_dummies(tips_df, drop_first=True)
```

# One hot encoding with scikit-learn

```python
# importing libraries
from sklearn.preprocessing import OneHotEncoder
```

```python
# creating object
oh_enc = OneHotEncoder(sparse=False, dtype='int64')
```

```python
# categorical variables in tips_df
cat_vars = tips_df.select_dtypes(include='O').columns
# we are applying the one hot encoder object on categorical variables od tips_df
oh_enc_arr = oh_enc.fit_transform(tips_df[cat_vars])
oh_enc_arr
```

```python
dummy_df.keys()
```

```python
# converting numpy array to data frame
ohc_enc_df = pd.DataFrame(oh_enc_arr, columns=['sex_Female', 'sex_Male', 'smoker_No',
       'smoker_Yes', 'day_Fri', 'day_Sat', 'day_Sun', 'day_Thur',
       'time_Dinner', 'time_Lunch'])
ohc_enc_df
```
