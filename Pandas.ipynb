```python
"""
Stack function is used to stack the prescribed levels from
columns to index and it returns a reshaped data frame or series having a m
multi level index with one or more new innermost levels compared to current data frame
"""
```

```python
my_tuple = list(zip(*[[1, 2, 3, 4, 5, 17, 18, 19], [11, 12, 13, 6, 7, 8, 9, 10]]))
```

```python
idx = pd.MultiIndex.from_tuples(my_tuple, names=["First", "Second"])
```

```python
df = pd.DataFrame(np.random.randn(8, 2), index=idx, columns=["A", "B"])
```

```python
df
```

```python
# slicing with respect to rows in original data frame
df2 = df[:4]
```

```python
df2
```

```python
# stack operation
df2.stack()
```

```python
# unstack operation
df2.unstack()
```

```python
"""
Pivot table 
Levels in the pivt table is stored in multi index object
All the index are columns of result data frame
"""
```

```python
df = pd.DataFrame({
                      "A" : ['a', 'b', 'c', 'd'] * 3,
                      "B" : ['A', 'B', 'C'] * 4,
                      "C" : ['P', 'P', 'P', 'Q', 'Q', 'Q'] * 2,
                      "D" : np.random.randn(12),
                      "E" : np.random.randn(12)
                 })
```

```python
# time series example
dates = pd.date_range('20200301', periods = 100, freq = 'S')
# we have genrated a dates starting from 2020-03-01 00:00:00 to 2020-03-01 00:01:39
# 100 periods
# difference between periods is 1 Second
dates
```

```python
# we have generated a time series whose indices are date and time of dates list
#                                 whose values are random
time_series = pd.Series(np.random.randint(0, 500, len(dates)), dates)
```

```python
time_series
```

```python
# we have resamples our series
time_series.resample('5min').sum()
```

```python
dates = pd.date_range('20200301', periods = 5, freq = 'S')
ts = pd.Series(np.random.randn(len(dates)), dates)
ts
```

```python
# localizing the time series
ts_utc = ts.tz_localize("UTC")
ts_utc
```

```python
# converting time series
ts_utc.tz_convert("US/Eastern")
```

```python
# let the frequency be  1 minute
dates = pd.date_range('20200301', periods = 5, freq = 'M')
ts = pd.Series(np.random.randn(len(dates)), dates)
ts
```

```python
# categorical
df = pd.DataFrame({
                      'id' : [1, 2, 3, 4, 5, 6],
                      'grade' : ['A', 'B', 'C', 'B', 'A', 'F']
                 })
df
```

```python
df['Grade'] = df['grade'].astype("category")
df["Grade"]
```

```python
df["Grade"].cat.categories = ["Excellent", "Good", "Can do better", "Repeat"]
df["Grade"] = df["Grade"].cat.set_categories(["Very Good", "Bad", "very bad", "Good", "medium"])
```
