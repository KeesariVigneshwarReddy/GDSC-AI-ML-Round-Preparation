# Logistic Regression as an example

```python
"""
making x array which contains all the 4 values -> R&D Spend	
                                                  Administration
                                                  Marketing Spend
                                                  Profit
"""
# Creating a numpy array of data frame
x = df[["R&D Spend", "Administration", "Marketing Spend", "Profit"]].values
x
```

```python
# second array
y = df["State"].values
y
```

```python
# since this is supervised learning we are to split data into train and test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
# In our whole data 20% is test data and 80% is train data
```

```python
# Algorithm 1 -> Logistic regression
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression()
logmodel.fit(x_train, y_train)
# we have fitted the training model on train_data
```

```python
# prediction
y_pred = logmodel.predict(x_test)
print(y_pred)
print(y_test)
```

```python
# confusion matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```
